```python
fig, ax = plt.subplots()
sns.barplot(x=['berkley','dublin','fremont','hayward','livermore','los Gatos','Mountain view','oakland','san francisco','san jose','santa clara','stockton','Sunnyvale','Tracy'],y=df.loc[(df.propertyTypeDimension=='Single Family'),['city','rentZestimate']].groupby('city')['rentZestimate'].mean(),data=df)
def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        
        patch.set_width(new_value)

        
        patch.set_x(patch.get_x() + diff * .5)

plt.xticks(rotation=45)
change_width(ax, 1)
sns.set(rc={'figure.figsize':(10,10)})
```
